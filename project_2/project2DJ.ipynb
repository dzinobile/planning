```python
h = 50
w = 180
buffer = 2

map = np.zeros((h, w, 3), dtype=np.uint8)
map_limits = map.copy()
map_limits = cv2.cvtColor(map_limits, cv2.COLOR_BGR2GRAY)
map_limits = cv2.rectangle(map_limits, (0,0), (w,h), (255), buffer)
locations = np.where(map_limits == (255))
boundary_xy = set(zip(locations[1], locations[0]))





def add_boundary(inpt_x,inpt_y):
    boundary_xy.add((inpt_x,inpt_y))
    for x in range(inpt_x-buffer, inpt_x+(buffer+1)):
        for y in range(inpt_y-buffer, inpt_y+(buffer+1)):
            boundary_xy.add((x,y))


for y in range(17,35):
    # E
    for x in range(20,40):
        if 16 < y <= 20:
            if 21 < x <= 39:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 20 < y <= 23:
            if 21 < x <= 25:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 23 < y <=27:
            if 21 < x <= 39:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 27 < y <= 30:
            if 21 < x <= 25:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 30 < y <= 34:
            if 21 < x <= 39:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
    
    #N
    for x in range(40,60):
        if 41 < x <= 45:
            map[y,x] = (255,0,0)
            add_boundary(x,y)
        if 55 < x <= 59:
            map[y,x] = (255,0,0)
            add_boundary(x,y)
        if 45 < x <= 55:
            if (1.2*x)-38 < y <= (1.2*x)-32:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
    
    #P
    for x in range(60, 80):
        if 61 < x <=67:
            map[y,x] = (255,0,0)
            add_boundary(x,y)
        if 67 < x <= 79:
            if (((x-67)**2)/(12**2))+(((y-22)**2)/(6**2)) <= 1:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
    
    #M
    for x in range(80,100):
        if 81 < x <=85:
            map[y,x] = (255,0,0)
            add_boundary(x,y)
        if 85 < x <= 90:
            if ((9/5)*x)-137 < y <= ((9/5)*x)-129:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 90 < x <= 95:
            if ((-9/5)*x)+187 < y <= ((-9/5)*x)+195:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 95 < x <= 99:
            map[y,x] = (255,0,0)
            add_boundary(x,y)
    
    #6
    for x in range(100,120):

        if ((x-110)**2)+(y-28.5)**2 <= 5.5**2:
            map[y,x] = (255,0,0)
            map[y,x+20] = (255,0,0)
            add_boundary(x,y)
            add_boundary(x+20,y)
        if ((x-110.85)**2)+(y-18)**2 <= 2**2:
            map[y,x] = (255,0,0)
            map[y,x+20] = (255,0,0)
            add_boundary(x,y)
            add_boundary(x+20,y)
        if x-92.85 < y <=28.5:
            if 12.85**2 < (((x-121.35)**2)+((y-28.5)**2)) <= 16.85**2:
                map[y,x] = (255,0,0)
                map[y,x+20] = (255,0,0)
                add_boundary(x,y)
                add_boundary(x+20,y)

    #1
    for x in range(140,160):
        if 16 < y <= 20.62:
            if 144 < x <= 152:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 20.62 < y <= 30:
            if 148 < x <= 152:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
        if 30 < y <= 34:
            if 144 < x < 156:
                map[y,x] = (255,0,0)
                add_boundary(x,y)
  






```

```python
start = (5, h-5)
end = (w-5, 5)

open_list = []
#closed_list = set()
closed_list = []
heapq.heapify(closed_list)
heapq.heapify(open_list)

heapq.heappush(open_list, (0, start, start))


def move(node,direction):
    parent_cost, parent = node[0], node[2]
    if direction == "up":
        child_cost, child = parent_cost+1, (parent[0],parent[1]-1)
    elif direction == "down":
        child_cost, child = parent_cost+1, (parent[0], parent[1]+1)
    elif direction == "left":
        child_cost, child = parent_cost+1, (parent[0]-1,parent[1])
    elif direction == "right":
        child_cost, child = parent_cost+1, (parent[0]+1,parent[1])
    elif direction == "up_right":
        child_cost, child = parent_cost+1.4, (parent[0]+1,parent[1]-1)
    elif direction == "down_right":
        child_cost, child = parent_cost+1.4, (parent[0]+1,parent[1]+1)
    elif direction == "down_left":
        child_cost, child = parent_cost+1.4, (parent[0]-1,parent[1]+1)
    elif direction == "up_left":
        child_cost, child = parent_cost+1.4, (parent[0]-1,parent[1]-1)
    
    if child in boundary_xy:
        return
    
    for item in open_list:
        if item[2] == child:
            if item[0] > child_cost:
                open_list.remove(item)
            else:
                return
        
    
    heapq.heappush(open_list, (child_cost,parent,child))
    return 



while end not in boundary_xy:
    parent_node = heapq.heappop(open_list)
    boundary_xy.add(parent_node[2])
    heapq.heappush(closed_list,parent_node)
    
    for direction in ["up","up_right","right","down_right","down","down_left","left","up_left"]:
        move(parent_node,direction)

path = []
path_node = end
while path_node != start:
    for item in closed_list:
        if item[2] == path_node:
            path.append(item)
            #print(item)
            path_node = item[1]

path.sort()

```

```python
rainbow = [(0,0,255)]


for i in range(0,len(closed_list)):
    ratio = i / (len(closed_list))

    if ratio < 1/3:
        B= 255
        G = int(255*(ratio*3))
        R = 0
    elif ratio < 2/3:
        B = int(255*(2-3*ratio))
        G = 255
        R = int(255*(3*ratio-1))
    else:
        B = 0
        G = int(255*(3-3*ratio))
        R = 255
    rainbow.append((B,G,R))

```
