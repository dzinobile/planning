```python
import numpy as np
import random

# Initialize state
init_state = [2, 3, 7, 5, 8, 0, 1, 6, 4]

# Write initial state to file
with open("Nodes.txt", "w") as node:
    node.write(" ".join(map(str, init_state)) + "\n")

def writenode(state):
    """Append a new state to Nodes.txt"""
    with open("Nodes.txt", "a") as node:
        node.write(" ".join(map(str, state)) + "\n")

# Location mappings for 3x3 grid
location_list = [(i % 3, i // 3) for i in range(9)]

def locate_blank(state):
    """Finds the coordinates of the blank (0) space."""
    return divmod(state.index(0), 3)  # Returns (row, col)

def move(state, direction):
    """Returns a new state if a move is possible, otherwise returns False."""
    row, col = locate_blank(state)
    new_row, new_col = row, col

    if direction == "up" and row > 0:
        new_row -= 1
    elif direction == "down" and row < 2:
        new_row += 1
    elif direction == "left" and col > 0:
        new_col -= 1
    elif direction == "right" and col < 2:
        new_col += 1
    else:
        return False, state  # Invalid move

    new_state = state[:]
    blank_index, swap_index = row * 3 + col, new_row * 3 + new_col
    new_state[blank_index], new_state[swap_index] = new_state[swap_index], new_state[blank_index]

    return True, new_state

# Goal state
goal = [1, 4, 7, 2, 5, 8, 3, 6, 0]

# BFS-like search to find the goal state
nodes_info = [(0, 0, init_state)]  # (index, parent_index, state)
i, p = 1, 0

while True:
    parent_state = nodes_info[p][2]

    for direction in ["up", "down", "left", "right"]:
        valid, new_state = move(parent_state, direction)
        if valid:
            nodes_info.append((i, p, new_state))
            if new_state == goal:
                break
            i += 1
    else:
        p += 1
        continue
    break  # Break outer loop if goal is found
```
